# Position Augmentation

```python
import copy

text = train_fold.text.iloc[1]
entity = copy.deepcopy(train_fold.entities.iloc[1]) # Deep Copy the Original List
new_entity = copy.deepcopy(train_fold.entities.iloc[1])
```

```python
def change_text(text,entity,pos1,pos2):
    first_entity = entity[pos1]
    second_entity = entity[pos2]
    start1,end1,_ = entity[pos1]
    start2,end2,_ = entity[pos2]
    text1 = text[start1:end1]
    text2 = text[start2:end2]
    text = text[0:start1]+text2+text[end1:start2]+text1+text[end2:]
    return text

def change_position(entity,pos1,pos2):
    new_entity = copy.deepcopy(entity)
    for i in range(pos1,pos2):
        if i == pos1:
            start1,end1,text1 = new_entity[pos1]
            start2,end2,text2 = new_entity[pos2]
            new_entity[pos1][2] = text2
            new_entity[pos2][2] = text1
            new_entity[pos1][1] = end2 - start2+start1
            diff = (end2 - start2) - (end1 - start1)
            new_entity[pos2][0] = entity[pos2][0]+diff
            new_entity[pos2][1] = entity[pos2][0]+end1 - start1+diff
        else:
            new_entity[i][0] = entity[i][0]+diff
            new_entity[i][1] = entity[i][1]+diff
    return new_entity

def entity_crop_list(text,entity):
    final_list = []
    for i in entity:
        start,end,_ = i
        final_list.append(text[start:end])
    return final_list

def pos_augmentation(text,entity):
    if len(entity)>3:
        pos1 = random.randint(0,len(entity)//2)
        pos2 = random.randint(pos1+1,len(entity)-1)
    elif len(entity)==2:
        pos1 = 0
        pos2 = 1
    else:
        return text,entity
    new_text = change_text(text,entity,pos1,pos2)
    new_entity = change_position(entity,pos1,pos2)
    return new_text,new_entity
```

```python
for i in range(0,100):
    text = train_fold.text.iloc[i]
    entity = copy.deepcopy(train_fold.entities.iloc[i])
    if len(entity) > 2:
        new_text,new_entity = pos_augmentation(text,entity)
        crop_list1 = entity_crop_list(text,entity)
        crop_list2 = entity_crop_list(new_text,new_entity)
        crop_list1.sort()
        crop_list2.sort()
        if crop_list1 != crop_list2:
            print(text,entity,new_text,new_entity,crop_list1,crop_list2, sep="\n")
            print('-'*50)
```

# random Upper/Lower Case

```python
from random import choice
sentence = 'Hello World 23131'
def randomize_case(text,entity):
    for i in entity:
        start,end,typer = i
        if typer == "ID":
            new_text = text[start:end]
            new_text = (''.join(choice((str.upper, str.lower))(c) for c in new_text))
            text_list = list(text) 
            text_list[start:end] = list(new_text)
    return ''.join(text_list),entity
```

# Repitition Augmentation

```python
import copy
text = train_fold.text.iloc[0]
entity = copy.deepcopy(train_fold.entities.iloc[0])
```

```python
def change_position2(entity,ent_pos,new_position,text_position):
    new_entity = copy.deepcopy(entity)
    chosen_entity = copy.deepcopy(entity[ent_pos])
    diff = new_entity[ent_pos][1] - new_entity[ent_pos][0]
    new_entity.insert(new_position,chosen_entity)
    new_entity[new_position][0] = text_position+1
    new_entity[new_position][1] = text_position+1+diff
    diff += 2
    for i in range(new_position+1,len(entity)+1):
        start1,end1,text1 = new_entity[i]
        new_entity[i][0] += diff
        new_entity[i][1] += diff
    return new_entity

def get_index(entity,chosen_pos):
    for i in range(len(entity)):
        if chosen_pos< entity[i][0]:
            return i
    return i+1

def rep_augmentation(text,entity):
    all_positions = []
    for j in entity:
        start,end,_ = j
        all_positions.extend(list(range(start,end+1)))
    possible_pos = []
    for k in list(range(len(text))):
        if k not in all_positions:
            possible_pos.append(k)
    old_entity_pos = random.choice(list(range(len(entity))))
    chosen_pos = random.choice(possible_pos)
    new_entity_pos = get_index(entity,chosen_pos)
    start,end,ent = entity[old_entity_pos]
    crop_text = text[start:end]
    new_text = text[0:chosen_pos]+' '+crop_text+' '+text[chosen_pos:]
    new_entity = change_position2(entity,old_entity_pos,new_entity_pos,chosen_pos)
    return new_text,new_entity
```

```python
for i in range(0,100):
    text = train_fold.text.iloc[i]
    entity = copy.deepcopy(train_fold.entities.iloc[i])
    all_positions = []
    for j in entity:
        start,end,_ = j
        all_positions.extend(list(range(start,end+1)))
    possible_pos = []
    for k in list(range(len(text))):
        if k not in all_positions:
            possible_pos.append(k)
    old_entity_pos = random.choice(list(range(len(entity))))
    chosen_pos = random.choice(possible_pos)
    new_entity_pos = get_index(entity,chosen_pos)
    start,end,ent = entity[old_entity_pos]
    crop_text = text[start:end]
    new_text = text[0:chosen_pos]+' '+crop_text+' '+text[chosen_pos:]
    new_entity = change_position2(entity,old_entity_pos,new_entity_pos,chosen_pos)
    crop_list1 = entity_crop_list(text,entity)
    crop_list1.append(crop_text)
    crop_list2 = entity_crop_list(new_text,new_entity)
    crop_list1.sort()
    crop_list2.sort()
    #if crop_list1 != crop_list2:
    print(crop_list1,crop_list2,new_text, sep="\n")
    print('-'*50)
```

```python
1.choose entity
2.choose list of special chars
3.add special chars to text
4.move entity start and end post addition
```

```python
chars = '/-@*.,!'
chars_list  = list(chars)
```

```python
def random_insert(special_char,text):
    list_text = list(text)
    list_text.insert(random.randint(0,len(list_text)),special_char)
    return ''.join(list_text)
```

```python
import copy
text = train_fold.text.iloc[1]
entity = copy.deepcopy(train_fold.entities.iloc[1])
print(text,entity)
```

```python
def special_chars_aug(text,entity):
    new_entity = copy.deepcopy(entity)
    cnt = 0
    for i in new_entity:
        start,end,typer = i
        if typer == "ID":
            new_text = random_insert(random.choice(chars_list),text[start:end])
            final_text = text.replace(text[start:end],new_text)
            cnt = 1
            i[1] += cnt
            continue
        i[0] += cnt
        i[1] += cnt
    print(new_entity)
    return final_text,new_entity
```
